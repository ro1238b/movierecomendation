```python
def extract_names(x):
    # expects a string representation of a list of dicts, returns list of 'name' values
    try:
        return [i['name'] for i in ast.literal_eval(x)]
    except Exception:
        return []

def get_top_cast(x, n=3):
    # expects a string representation of a list of dicts, returns list of top n 'name' values
    try:
        cast_list = ast.literal_eval(x)
        return [i['name'] for i in cast_list[:n]]
    except Exception:
        return []

for col in ['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages']:
    movies[col] = movies[col].apply(lambda x: extract_names(x) if isinstance(x, str) else x)

credits['cast'] = credits['cast'].apply(lambda x: get_top_cast(x) if isinstance(x, str) else x)
credits['director'] = credits['crew'].apply(lambda x: get_director(x) if isinstance(x, str) else x)

# Merge datasets
df = movies.merge(credits[['title', 'cast', 'director']], on='title')

# Preview
print(df[['title', 'genres', 'keywords', 'cast', 'director']].head())
```

```python
def clean_text(x):
    return ' '.join(x) if isinstance(x, list) else ''

df['tags'] = df['genres'].apply(clean_text) + ' ' + \
             df['keywords'].apply(clean_text) + ' ' + \
             df['cast'].apply(clean_text) + ' ' + \
             df['director'].apply(lambda x: x if isinstance(x, str) else '')

df['tags'] = df['tags'].str.lower()
```

```python
%pip install tf-keras

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')  # Fast and accurate for semantic similarity

# Fill NaNs and convert tags to list of strings
df['tags'] = df['tags'].fillna('')

# Generate BERT embeddings
embeddings = model.encode(df['tags'].tolist(), show_progress_bar=True)

# Compute similarity
similarity = cosine_similarity(embeddings)
```

```python
def recommend(movie_title):
    if movie_title not in df['title'].values:
        print("Movie not found.")
        return
    
    index = df[df['title'] == movie_title].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    print(f"\nTop 5 recommendations for: {movie_title}")
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```
